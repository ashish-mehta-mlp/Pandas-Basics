## How do I filter Rows of Pandas dataframe by Column Value??

```python
import pandas as pd
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies.head()
```

```python
movies.shape
```

```python
# In this case, we will be filtering the data using "duration" series. We will filter and use only those rows whose duration
# is atleast 200 or more
```

```python
# We are going to create a python list of boolean that has the same exact length as the data frame
# The boolean will be True, if the given row has duration <= 200 else it will be False.
```

```python
boolean = []
for x in movies.duration:
    if x >= 200:
        boolean.append(True)
    else:
        boolean.append(False)
```

```python
boolean[0:5]
```

```python
len(boolean)

#hence, we have created a list of booleans that has the same length as the dataframe and effectively tells us, which rows match
#the conditions
```

```python
# Step 2: We will convert a Booleans list into a Pandas Series.
is_long = pd.Series(boolean)
```

```python
is_long.head()
```

```python
# Step 3: Now we will pass the is_long series to the data_frame movies using bracket notation.
movies[is_long]

#This has filtered the dataset only with movies greater than 200.
```

## Method 2 of Filtering

```python
is_long_as = movies.duration < 200 # This line replaces the entire for loop. It gives the boolean value for entire duration column/series
is_long_as.head() 
```

```python
movies[is_long_as] # This does not change the actual dataset as inplace is not used
```

```python
movies.sort_values("duration")
```

## Method 3 to filter

```python
movies[movies.duration >=220] #This line of code tells pandas which rows we want to select.
```

## Video 8: Bonus

After applying a filter, what if we want to study only one particular series say 'genre' of the movies?

```python
movies[movies.duration >= 200]['genre'] # In this case, we have filtered a dataset using a condition (duration > 200) and then pulled out a desired column 
```

```python
# Instead of using 2 seperate [], we can use .loc method which is very powerful

movies.loc[movies.duration >= 225, 'genre']
```

## How do I filter rows using Multiple Conditions?? 

```python
import pandas as pd
```

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies.head()
```

```python
movies[movies.duration >= 200]
```

OR operator And operator

OR operator is True if any one of the side is True
True OR False = True

AND operator is True only if both sides are True
True AND True = True

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')] 

# First we use the [and] operator
# Then we put parentheses() for each condition so that pandas can understand the order of filtering. First >= 200 then Drama
# Instead of using and operator, we need to use & sign.
```

```python
# AND == &
# OR == |
```

## Video 14: When Should I use "groupby" in pandas?

```python
import pandas as pd
```

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
drinks.head()
```

```python
# What is the average beer servings of all countries combined?
drinks.beer_servings.mean()
```

```python
## Now we have to calculate continent wise beer_servings
drinks.groupby("continent").beer_servings.mean()
```

```python
drinks[drinks.continent == "Africa"] # Filtering the dataframe with only africa as Continent
```

```python
drinks[drinks.continent == "Africa"].beer_servings.mean()
```

```python
# WE SHOULD USE GROUPBY WHEN WE ARE ANALYSING A PANDAS SERIES WITH SOME CATEGORY
# IN OUR EXAMPLE, THE CATEGORY IS CONTINENT AND WE ARE ANALYZING CONTINENT VISE BEER SERVINGS
# IF WE CAN REPHRASE OUR QUESTION AS "FOR EACH" THEN IT IS A GOOD CANDIDATE OF GROUPBY. 
# EG: FOR EACH CONTINENT, WHAT IS THE AVERAGE BEER SERVING?
```

```python
drinks.groupby("continent").beer_servings.max()
```

```python
drinks.groupby("continent").beer_servings.min()
```

```python
# The more powerful method than mean, min, max is .agg
# .agg helps us to specify multiple aggrigation functions such as min,max,mean etc at once
```

```python
drinks.groupby("continent").beer_servings.agg(["count","min","max","mean"])
```

## Video 14: Bonus

it will calculate the operation (mean,min,max etc) on all the possible columns

## Video 16: How do I handle missing values in Pandas

```python
import pandas as pd
```

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.tail()
```

```python
## NaN: means a missing value.
## When building the dataframe, read_csv detected that there is a missing value and filled it with NaN (Not a number)
```

```python
## Various Methods to work with missing values

## Method 1: .isnull [Dataframe method]

ufo.isnull().tail()
## The way isnull works is that it will show False if the value is not null and will show True if the value is null(NaN)
## It is only because pandas uses this NaN notation that isnull is able to detect True and False for null values
```

```python
## There is an inverse of isnull() called notnull()

ufo.notnull().tail()
```

```python
## Why are isnull() and notnull() useful?

ufo.isnull().sum() # This tells us the number of missing values in each column
```

```python
#Explanation of above command:

pd.Series([True, False, True, True]).sum()
# When we take the sum of boolean, the pandas converts True's to 1 and False's to 0 and then adds them up.
```

```python
## Note that isnull is also a series method. And can be used to identify the subset of Null values.

ufo[ufo.City.isnull()]
```

```python
ufo.shape
```

```python
## What should we do with the missing values??

## 1: We can drop the NaN values(dropna)

ufo.dropna(how='any').shape #We want pandas to drop the row if any of its value are NaN
                            # inplace = False by default
```

```python
ufo.dropna(how='all').shape #We want pandas to drop a row if all of its values are NaN
```

```python
ufo.dropna(subset=["City", "Shape Reported"], how = "any").shape #We want pandas to drop the row if the value of City or Shape Reported is NaN
```

```python
ufo.dropna(subset=["City", "Shape Reported"], how = "all").shape #We want pandas to drop the row if the value of City And Shape Reported both are NaN
```

```python
movies.content_rating.isnull().sum()
```

SO WHAT EXACTLY IS GOING ON WITH THE WARNING? (SettingWithCopy)

movies[movies.content_rating == "NOT RATED"].content_rating = np.nan

This line of code is actually two operation. Portion till 'NOT RATED'] is known as get_item. The part after that is set_item.
Pandas cant guarentee that whether the first part of the code returned a view or a copy of the data. If it returned view, 
then set_item would affect the dataframe. But if it returned copy, the set_item would make some modifications in the copy but 
the copy would immediately get discarded and thus the original dataframe does not get modified.
loc solves this problem by converting it from 2 operations to single operation code. That is why it does not throw a warning

Summary: if we are trying to select rows and columns in the same line of code, use .loc

```python
#Eg 2: We are working with movies that have a very high star rating

top_movies = movies.loc[movies.star_rating >= 9, :]
```

```python
top_movies
```

```python
# Lets assume that we come to know that duration of Shawshank Redemption is wrong and we want to fix it

top_movies.loc[0, 'duration'] = 150
```

```python
# We are using .loc but still it is showing a SettingWithCopyWarning
```

```python
top_movies
```

If the code worked, then what was the problem?

Pandas actually isnt sure whether top_movies is a view or a copy of main movies dataset. So it warns us asking if we are 
modifying just one thing ie just top_movies or are we modifying 2 things (top_movies and movies)

# The solution is this:

top_movies = movies.loc[movies.star_rating >= 9, :]

The problem was this line of code. Anytime we try to create a dataframe copy, we should explicitly use the .copy() method 
so that pandas can be sure that it is a copy of the dataframe. 

```python
top_movies = movies.loc[movies.star_rating >= 9, :].copy()
```

## Video 27: How do I make my Pandas dataframe Smaller and Faster??

```python
import pandas as pd
```

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
drinks.head()
```

```python
drinks.info()
```

```python
# there is a "memory usage" section with a + sign. It tells us that the dataframe takes atleast 9 kb of memory
# The obeject columns are a reference to other objects stored some place else. Pandas wants to run the info method quickly
# It does not want the info method to go and search for objects and find their memory info. Hence it uses + sign 
```

```python
# We can ask pandas to calculate the actual memory usage

drinks.info(memory_usage = 'deep')

# The actual memory usage is 30.4kb
```

```python
#How much space is each column taking?

drinks.memory_usage() #It gives us the memory usage in bytes and not KB

#These are not the real memory usage per column. Just like drinks.info() it does not deeply inspect the object columns 
```

```python
drinks.memory_usage(deep = True) # To get the actual values, we put deep = True
```

```python
# As the result obtained from drinks.memory_usage() is a series with column names as index, we can actually take 
# the sum of all the integers

drinks.memory_usage(deep = True).sum() #Sum comes down to 31 KB
```

Object series can take up alot of data space. So how can we be more space effiecent especially with object columns given 
that the data set is very large.


Sol 1: We can store our strings as integers.
    For eg: We have a column of continents. There are only 6 unique values in this column which are 6 continents. So we 
        can assign an int value to each continent say: Africa : 0, Asia : 1, Australia : 2 and so on and store that column as 
        string. 
        If we invented this system then we have to create a lookup table saying 0 means Africa, 1 means Asia and so on. So in a
        way we still have to store the strings but we only have to store it once.
        One great advantage of Pandas is that Pandas itself creates a lookup table for us.

```python
drinks.continent.head()
```

```python
# How to convert an object type column into a category type?

drinks["continent"] = drinks.continent.astype("category")
```

```python
drinks.dtypes # The country is still an object but continent has now become a category
```

```python
drinks.continent.head()

# It still looks the same here. but notice that pandas has created a list of 6 categories and under the hood it is storing 
# each continent name as an integer
```

```python
drinks.continent.cat.codes.head() # Proof that continents are stored as numbers
```

```python
drinks.memory_usage(deep = True)

# Instead of storing 193 strings, we are storing a 193 integers and hence the memory usage has gone down drastically 
# for continent column
```

```python
drinks.memory_usage(deep = True).sum() 
```

```python
# Lets try to repeat this for other object series that is "Country"
import pandas as pd
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
```

```python
drinks["country"] = drinks.country.astype("category")
```

```python
drinks.memory_usage(deep = True)

# When we converted country series to category, got larger. Why is that?? 
# The reason is we are creating 193 unique integers for 193 unique countries plus a lookup table with 193 integers pointing to
# 193 strings. Hence the space has increased instead of decreasing.
```

Bottom line is we should convert an dtype object to category only when we have a series with strings that are repeating again 
and again.

Besides reducing the memory usage, it also speeds up our computations.

apply is both a series method and a dataframe method. It applies a function to each element in a series. 

```python
# Lets pretend, we have to calculate length of each of the strings in the name column and create a new column called name_len
# that contains the integer value.
# 1. Used as a series method

train["Name_len"] = train.Name.apply(len)

# Hence apply function when used as a series method applies len function of Name series
```

```python
# We have to extract last name of each person to its own column
# We that we need to get the part before comma.

train.Name.str.split(',').head()
```

```python
# The Name column in train dataset and the o/p of above command look similar. But there is a huge difference.
# The values of Name column is a string. But the values of the above command is a list of string. Each value of row is a list of
# string [Braund, Mr. Owen Harris] These are list of 2 string values
```

```python
# Now all we want is the first part of each list.
# We have created a function to extract the fist part before , and will apply it below

def get_element(my_list, position):
    return my_list[position]
```

3. Applymap

applymap is also a dataframe method. Applymap applies a function to every element of the dataframe. It does not go along horizontal or vertical axis like apply function for dataframe. But it is applied to each and every element.

```python
drinks.loc[:, "beer_servings":"wine_servings"].applymap(float).head()
```

Q2. Why is it that some of the commands have an inclusive range (.loc) and some of the commands have exclusive range (.iloc) in Pandas?

Ans. Pandas is built on numpy. And most of the time, pandas tries to follow the convention of numpy. So iloc is exclusive of last number. And thats how numpy does its slicing. Hence iloc tries to follow the convention of numpy. And why does numpy do that? Its because thats how pythons range does it. 
Then why doesnt the loc do the same thing? Why is loc inclusive of the last value? it makes sense because when the coder reads it, he would not know which column label comes first. If we put .loc[:, "City":"Time"] and if the result shows columns till "State" only, then coder would not know that state comes before time and he might get confused. Hence loc is inclusive on both sides.
